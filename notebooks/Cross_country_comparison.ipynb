```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
benin= pd.read_csv('../data/benin-malanville.csv')
sl= pd.read_csv('../data/sierraleone-bumbuna.csv')
togo= pd.read_csv('../data/togo-dapaong_qc.csv')
```

```python
benin['Country'] = 'Benin'
togo['Country'] = 'Togo'
sl['Country'] = 'Sierraleone'
```

```python
combined = pd.concat([benin, togo, sl])
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Country', y='GHI', data=combined)
plt.title('GHI Distribution by Country')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Country', y='DNI', data=combined)
plt.title('DNI Distribution by Country')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Country', y='DHI', data=combined)
plt.title('DHI Distribution by Country')
plt.show()
```

```python
summary = combined.groupby('Country')[['GHI', 'DNI', 'DHI']].agg(['mean', 'median', 'std']).round(2)
print(summary)
```

```python
""" 
Key Observation:
- Out of the three, Benin has the highest GHI, while Sieraleone Has the lowest
- There std ranges between 156.52 and 158.95   
- Benin and Sierraleone have a -0.1 DHI median while Togo has 0
"""
```

```python
avg_ghi = combined.groupby('Country')['GHI'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=avg_ghi.index, y=avg_ghi.values)
plt.title('Average GHI by Country')
plt.ylabel('Average GHI (W/m)')
plt.show()
```
