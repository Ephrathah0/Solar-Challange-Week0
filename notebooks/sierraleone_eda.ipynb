```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../data/sierraleone-bumbuna.csv')
```

```python
df.info()
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
plt.figure(figsize=(10,5))
plt.plot(df['Timestamp'], df['GHI'])
plt.title('Global Horizontal Irradiance Over Time - Sierraleone')
plt.xlabel('Time')
plt.ylabel('GHI (W/m)')
plt.show()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
plt.figure(figsize= (10,5))
plt.title('Direct Normal Irradiance Over Time - Sierraleone')
plt.plot(df['Timestamp'], df['DNI'])
plt.xlabel('Time')
plt.ylabel('Direct Normal Irradiance Over Time')
plt.show()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
plt.figure(figsize=(10,5))
plt.plot(df['Timestamp'], df['DHI'])
plt.title('Diffuse Horizontal Irradiance Over Time - Sierraleone')
plt.xlabel('Time')
plt.ylabel('DHI (W/m)')
plt.show()
```
